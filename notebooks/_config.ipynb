## 共通設定
`config.yaml` から設定を読み込みます。各notebookの先頭で `%run ./_config` を実行してください。

```python
import yaml, os

# config.yaml の読み込み
try:
    # Databricks Workspace 上での実行
    _nb_path = dbutils.notebook.entry_point.getDbutils().notebook().getContext().notebookPath().get()
    _cfg_dir = os.path.dirname(os.path.dirname(_nb_path))
    _cfg_path = f"/Workspace{_cfg_dir}/config.yaml"
    with open(_cfg_path) as f:
        cfg = yaml.safe_load(f)
except Exception:
    # ローカル実行時のフォールバック
    with open("../config.yaml") as f:
        cfg = yaml.safe_load(f)

# Unity Catalog
catalog = cfg["catalog"]
schema = cfg["schema"]
volume_name = cfg["volume_name"]

# Vector Search
VECTOR_SEARCH_ENDPOINT_NAME = cfg["vector_search_endpoint_name"]
VECTOR_SEARCH_INDEX_NAME = cfg["vector_search_index_name"]
EMBEDDING_MODEL_NAME = cfg["embedding_model_name"]

# LLM
LLM_ENDPOINT_NAME = cfg["llm_endpoint_name"]
AI_GATEWAY_URL = cfg["ai_gateway_url"]

# MLflow
MLFLOW_EXPERIMENT_NAME = cfg.get("mlflow_experiment_name", "")
OBS_SCHEMA = cfg.get("obs_schema", "dify_observability")
DELTA_SYNC_TABLE = cfg.get("delta_sync_table", "mlflow_trace_log_delta_sync")

# Genie
GENIE_SPACE_ID = cfg.get("genie_space_id", "")

# Workspace情報（Databricksランタイムから取得）
try:
    host = spark.conf.get("spark.databricks.workspaceUrl")
    token = dbutils.notebook.entry_point.getDbutils().notebook().getContext().apiToken().get()
except Exception:
    host = os.environ.get("DATABRICKS_HOST", "")
    token = os.environ.get("DATABRICKS_TOKEN", "")

# 確認
print(f"{'='*50}")
print(f"Catalog:          {catalog}")
print(f"Schema:           {schema}")
print(f"Volume:           {volume_name}")
print(f"Workspace:        {host}")
print(f"{'='*50}")
print(f"LLM Endpoint:     {LLM_ENDPOINT_NAME}")
print(f"AI Gateway URL:   {AI_GATEWAY_URL}")
print(f"VS Endpoint:      {VECTOR_SEARCH_ENDPOINT_NAME}")
print(f"VS Index:         {VECTOR_SEARCH_INDEX_NAME}")
print(f"Embedding Model:  {EMBEDDING_MODEL_NAME}")
print(f"{'='*50}")
print(f"MLflow Experiment: {MLFLOW_EXPERIMENT_NAME}")
print(f"Obs Schema:       {OBS_SCHEMA}")
print(f"Delta Sync Table: {DELTA_SYNC_TABLE}")
print(f"Genie Space ID:   {GENIE_SPACE_ID or '(未設定)'}")
```
