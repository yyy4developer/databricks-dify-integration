# Pattern ⑤ Databricksオーケストレーター

> **※ 本パターンは紹介のみです（未検証）**

## 概要

Pattern ①〜④は **Dify → Databricks** の方向でしたが、  
Pattern ⑤は逆方向: **Databricks → Dify API** を呼び出すパターンです。  
大規模データに対する**バッチAI処理**（要約・分類・感情分析）に適しています。

## アーキテクチャ

```
Databricks Notebook / Workflow
  │
  ├── Spark: テーブルから大量データを読み込み（数千〜数万件）
  │
  ├── 各レコードに対して Dify Workflow API を呼び出し
  │     POST /v1/workflows/run（blocking）
  │     └── Difyが分類・要約・感情分析を実行
  │
  └── 結果をDeltaテーブルに書き戻し → ダッシュボード / 分析
```

## 他パターンとの使い分け

| 方向 | パターン | ユースケース |
|------|---------|----------|
| Dify → Databricks | ①②③④ | ユーザー対話型（チャット、Q&A） |
| **Databricks → Dify** | **⑤** | **大規模バッチ処理（分類、要約、抽出）** |

## 注意点

- DifyをローカルDockerで動かしている場合、Databricksからアクセスできない  
  → Dify Cloud（SaaS）またはパブリックにデプロイされたDifyが必要
- Dify APIのレート制限に注意（大量リクエスト時は並列度を制御）
